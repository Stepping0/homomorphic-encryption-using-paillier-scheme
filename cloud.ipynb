since we do not have private keys, we can not decrypt the data

Paillier cryptosystem is additively homomorphic. We can perform operations on encrypted data in the cloud without performing any decryption operation.

```python
c3= c1+c2
```

```python
c3
```

We can not perform decryption operation on cloud side since we do not have the private key.

```python
cs.decrypt(c3)
```

c3 is sent to onprem side to decrypt.

We can not do decryption but we can perform enryption operation on cloud side.

```python
bonus=cs.encrypt(1000)
```

```python
bonus
```

```python
c4=c1+bonus
```

```python
c4
```

Although the Paillier cryptosystem is not multiplicatively homomorphic, it allows the ciphertext to be multiplied by a constant

```python
c5= c1*0.5
```

```python
c5
```

In the Paillier cryptosystem, the cipher text can be recreated. In other words, a plaintext can have more than one encrypted version.

```python
c3_prime=cs.regenerate_ciphertext(c3)
```
