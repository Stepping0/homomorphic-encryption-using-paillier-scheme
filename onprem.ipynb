In this code i'll cover onprem/server side for paillier cryptosystem with homomorphic ecnryption

on onprem side we have both private and public keys

```python
from lightphe import LightPHE
```

```python
cs=LightPHE(algorithm_name="Paillier")
```

```python
cs.cs.keys
```

```python
cs.export_keys(target_file="public.txt",public=True)
```

```python
m1=10000 #maas
c1=cs.encrypt(m1)
```

```python
c1
```

```python
m2=2000
c2=cs.encrypt(m2)
```

```python
c2
```

```python
cs.decrypt(c1)
```

```python
assert cs.decrypt(c1)==m1
```

```python
cs.decrypt(c2)
```

```python
assert cs.decrypt(c2)==m2
```
